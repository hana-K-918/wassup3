```python
import requests
from bs4 import BeautifulSoup
```

```python
PATH = 'https://finance.naver.com/'
resp = requests.get(PATH)
resp
```

```python
src = resp.text
```

```python
soup = BeautifulSoup(src, 'lxml')
print(soup)
```

```python
srclist = soup.select('.section_strategy')
print(srclist)
```

```python
srclist = soup.select('.section_strategy li a')
print(srclist)
```

```python
for a in srclist:
    print(a.text)
```

```python
url = srclist[0]['href']
PATH + url
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, url)
```

```python
news_title = []
news_url = []
for a in srclist:
    title = a.text
    url = urljoin(PATH, a['href'])
    news_title.append(title)
    news_url.append(url)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

```python
df.to_excel('output/0622_review.xlsx', index=False)
```
